```python
from skimage import io, data

image = data.coins()

io.imshow(image)
```

# Read Image di Dalam Folder

```python
img_path = "image_original/" #path folder image anda
image = io.imread(img_path + '20200901-160150.jpg') #baca image
io.imshow(image)  #tampilkan image
```

```python
from skimage.color import rgb2gray #library convert dari RGB image ke GreyScale Image
image_gray = rgb2gray(image) #convert dari RGB image ke GreyScale Image
io.imshow(image_gray)
io.show()
```

```python
image.shape #cek berapa width, height, dan dimensi dari pixel
#image
```

```python
image_gray.shape #cek berapa width, height, dan dimensi dari pixel
#image_gray
```

```python
io.imshow(image_gray)
io.show()
```

## Crop image sesuai koordinat yang diinginkan 

### Yang mau di crop pada image ini yaitu tanaman yang diatas lakban hitam (background hitam)

```python
#array pertama adalah koordinat vertikal yang ingin dicrop 
#array kedua adalah koordinat horizontal yang ingin dicrop

image_gray_crop = image_gray[380:480, 310:400]
io.imshow(image_gray_crop)
io.show()
```

# Thresholding image (mengubah ke binary image)

```python
from skimage import filters

nilai_thershold = [] #nilai threshold yang akan ditampung

nilai_thershold = filters.threshold_li(image_gray_crop, iter_callback=nilai_thershold.append)

#hasil image yang ingin diambil adalah nilai image yang lebih besar dari nilai threshold
binary_image = image_gray_crop > nilai_thershold

#hasil threhsold image menggunakan image yang sudah dicrop
io.imshow(binary_image)
```

```python
from skimage import filters

nilai_thershold = []

nilai_thershold = filters.threshold_li(image_gray, iter_callback=nilai_thershold.append)

binary_image = image_gray > nilai_thershold
#hasil threhsold image menggunakan image yang belum dicrop
io.imshow(binary_image)
```

# Resize Image

```python
from skimage.transform import rescale, resize, downscale_local_mean

image_resized = resize(image_gray_crop, (200,200), #resize image menjadi 200 x 200 pixels
                       anti_aliasing=True)

io.imshow(image_resized)
```

```python
image_resized.shape #image yang sudah di resized, width dan height nya sudah 200 x 200 pixels
```

```python
image_gray_crop.shape #image yang sudah di resized, width dan height nya sudah 80 x 100 pixels
```

# Read semua image di dalam folder

```python
import os

img_path = 'image_original/' #path folder image

list_files = os.listdir(img_path)
list_files
```

## Siapkan dulu folder untuk menyimpan image yang akan di resized dan di crop

```python
img_resize_path = 'image_resized/' #path folder image resized
img_crop_path = 'image_crop/' #path folder image crop

#image_list = []

for filename in list_files:
    input_image = io.imread(img_path + filename) #read seluruh image yang ada di folder "image_original/"
    
    #Proses resizing semua image di dalam folder "image original"
    image_resized = resize(input_image, (200,200), anti_aliasing=True)
    
    #Proses croping semua image di dalam folder "image original"
    image_crop = input_image[400:480, 300:400]
    
    io.imsave(img_crop_path + filename, image_crop) #save image yang sudah di resize ke dalam folder "image_resized/"
    
    io.imsave(img_resize_path + filename, image_resized) #save image yang sudah di crop ke dalam folder "image_crop/"
    
```
